```python
import operator
def resolve_saves(value):
    result = []
    hits, saves = value
    normal_hits = operator.countOf(hits, 'Success')
    crit_hits = operator.countOf(hits, 'Crit')
    normal_saves = operator.countOf(saves, 'Success')
    crit_saves = operator.countOf(saves, 'Crit')
    crit_hits = max(0, crit_hits - crit_saves)
    hits = max(0, normal_hits - normal_saves - max(0, crit_saves - crit_hits))
    result += ['Crit']*crit_hits + ['Success']*hits
    return (tuple(result), crit_hits)
def ignore_old_crits(value):
    hits, _ = value
    return hits
roll=lea.pmf({'Miss': 2/6, 'Success': 3/6, 'Crit': 1/6})
saved = lea.joint(roll.draw_sorted_with_replacement(4), save3.draw_sorted_with_replacement(1)).map(resolve_saves).map(ignore_old_crits)
saved.map(lambda x: 6*list(x).count('Crit') + 5*list(x).count('Success')).plot()
roll=lea.pmf({'Miss': 1/6, 'Success': 3/6, 'Crit': 1/6})
saved = lea.joint(roll.draw_sorted_with_replacement(4), save3.draw_sorted_with_replacement(2)).map(resolve_saves).map(ignore_old_crits)
saved.map(lambda x: 6*list(x).count('Crit') + 5*list(x).count('Success')).plot()
```

```python
def kt_roll(amount, target, crit_value=6):
    return lea.interval(1,6).map(
        lambda x: 'Crit' if x >= crit_value else 'Success' if x>= target else 'Miss' 
    ).draw(amount, sorted=True, replacement=True)

class KtRoll(object):
    
    def __init__(self, target, crit_value=6):
        self.base_roll = lea.interval(1,6).map(
            lambda x: 'Crit' if x >= crit_value else 'Success' if x>= target else 'Miss' 
        )
    
    def roll(self, amount, rerolls=0):
        return lea.joint(
            self.base_roll.draw(amount, sorted=True, replacement=True),
            self.base_roll.draw(rerolls)
        ).map(self.map_rerolls)

    def map_rerolls(self, value):
        hits = list(value[0])
        for reroll in list(value[1]):
            if reroll == 'Crit' and 'Miss' in hits:
                hits.remove('Miss')
                hits.insert(0, 'Crit')
                continue
            if reroll == 'Success' and 'Miss' in hits:
                hits.remove('Miss')
                hits = [x for x in hits if x == 'Crit'] + ['Success'] + [x for x in hits if x != 'Crit']
                continue
        return tuple(hits)

class Target(object):
    
    def __init__(self, df, sv):
        self.df = df
        self.sv = sv
    
    def save(self):
        return KtRoll(self.sv)
    
    def get_defense(self):
        return self.df
    
    def __str__(self):
        return 'Sv: {} Df: {}'.format(self.sv, self.df)
    
class Weapon(object):
    
    ap = 0
    mw = 0
    
    def __init__(self, name, attacks, bs, nd, cd, special_rules=None, crit_rules=None):
        self.name = name
        self.attacks = attacks
        self.bs = bs
        self.nd = nd
        self.cd = cd
        self.special_rules = special_rules or []
        if 'AP2' in self.special_rules:
            self.ap = 2
        if 'AP1' in self.special_rules:
            self.ap = 1
        if 'MW3' in self.special_rules:
            self.mw = 3
        self.crit_rules = crit_rules or []
    
    def plot_shoot(self, *kargs, **kwargs):
        desc = ''
        if 'attack_rerolls' in kwargs:
            desc += ' attacker rerolls: {}'.format(kwargs['attack_rerolls'])
        if 'save_rerolls' in kwargs:
            desc += ' save rerolls: {}'.format(kwargs['save_rerolls'])
        title = '{} vs {} {}'.format(self.name, kargs[0], desc)
        self.shoot(*kargs, **kwargs).plot(title=title)
    
    def shoot(self, target, cover=None, attack_rerolls=0, save_rerolls=0):
        return lea.joint(
            KtRoll(self.bs).roll(self.attacks, attack_rerolls), 
            target.save().roll(target.get_defense() - self.ap, save_rerolls)
        ).map(resolve_saves).map(self.damage)
    
    def damage(self, results):
        hits, og_crits = results
        return self.cd*list(hits).count('Crit') + self.nd*list(hits).count('Success') + og_crits*self.mw

    

kt_roll(1,4)
#apply_damage(KtRoll(4).roll(3, rerolls=1),2, 3).plot()
#apply_damage(KtRoll(3).roll(4, rerolls=1),5, 6).plot()

plasma = Weapon('Plasma - Supercharge', 4, 3, 5, 6, ['AP2', 'Hot'])
plasma.plot_shoot(Target(3,3), attack_rerolls=0)
```

```python
# library
import seaborn as sns
import pandas as pd
import numpy as np
import itertools
 
# Create a dataset
#df = pd.DataFrame(matrix, columns=["a","b","c","d","e"])
 
# Default heatmap: just a visualization of this square matrix
#sns.heatmap(df)

class WeaponDamageProfile(object):
    
    def __init__(self, weapon):
        self.weapon = weapon
        
    def generate_matrix(self):
        target_probs = []
        columns = []
        for target in [Target(3,3), Target(3,4), Target(3,5)]:
            for save_reroll, attack_reroll in itertools.product(range(1,-1, -1), range(3)):
                base = [0]*30
                for (wounds, prob) in list(self.weapon.shoot(target, save_rerolls=save_reroll, attack_rerolls=attack_reroll).pmf_tuple):
                    base[wounds] = prob
                target_probs.append(base)
                columns.append(str(target) + ' dRR:{} aRR:{}'.format(save_reroll, attack_reroll))
        return (target_probs, columns)

def plot_heatmap(weapon):
    probs, columns = WeaponDamageProfile(weapon).generate_matrix()
    df = pd.DataFrame(np.transpose(probs), columns=columns)
    ax = sns.heatmap(df)
    ax.invert_yaxis()
    ax.set_title(weapon.name)

plot_heatmap(plasma)
```
