```python
from ktstats import Fight, MeleeWeapon, Target

def describe_combat_result(x):
    return '{} ({:0>2}) vs {} ({:0>2})'.format(x[0][0], x[0][1], x[1][0],x[1][1])

probs = Fight(
    Target(3,4, wounds=10, name='Rusty'),
    MeleeWeapon('Chordclaw and transonic razor', 5, 3, 4, 5, ['Balanced']),
    Target(3,4, wounds=8, name='Shade Runner'),
    MeleeWeapon('Hekatarii blades', 4, 3, 3, 5, ['Lethal 5+', 'Relentless'])
).fight().map(describe_combat_result)
print(probs.as_string('%-',nb_decimals=2))
```

```python
fight = Fight(
    Target(3,4, wounds=10, name='Rusty'),
    MeleeWeapon('Transonic Blades', 5, 3, 4, 6, ['Rending']),
    Target(3,4, wounds=8, name='Shade Runner'),
    MeleeWeapon('Hekatarii blades', 4, 3, 3, 5, ['Lethal 5+', 'Relentless'])
)
print ('no additional rerolls')
probs = fight.fight().map(describe_combat_result)
print(probs.as_string('%-',nb_decimals=2))
print ('attacker rerolling 1')
probs = fight.fight(1).map(describe_combat_result)
print(probs.as_string('%-',nb_decimals=2))
```

```python
def survives_or_not(wounds, fighter):
    if wounds == 0:
        return 'dead'
    else:
        return 'survives'

def op_state(wounds, fighter):
    if wounds == 0:
        return 'dead'
    if wounds < int(fighter.wounds / 2):
        return 'injured'
    if wounds == fighter.wounds:
        return 'untouched'
    else:
        return 'survives'
    
    
def kill_or_not(result, attacker, defender, show_state=False):
    ar, dr = result
    func = op_state if show_state else survives_or_not
    return '(A) {} {} - (D) {} {}'.format(
        attacker.name,
        func(ar[1], attacker),
        defender.name,
        func(dr[1], defender)
    )


fight = Fight(
    Target(3,4, wounds=10, name='Rusty'),
    MeleeWeapon('Chordclaw and transonic razor', 5, 3, 4, 6, ['Balanced']),
    Target(3,4, wounds=7, name='Pathfinder'),
    MeleeWeapon('Fists', 3, 5, 2, 3, [])
)


def print_combat_results(combat_probs):
    print('Combat Results:\n--------------')
    print('Survibability:\n')
    probs = combat_probs.map(lambda x: kill_or_not(x, fight.attacker, fight.defender))
    print(probs.as_string('%-',nb_decimals=2))
    print('\nOperative state:\n')
    probs = combat_probs.map(lambda x: kill_or_not(x, fight.attacker, fight.defender, True))
    print(probs.as_string('%-',nb_decimals=2))
    print('\nFull Probs:\n')
    probs = combat_probs.map(describe_combat_result)
    print(probs.as_string('%-',nb_decimals=2))

print_combat_results(fight.fight(1))
```
